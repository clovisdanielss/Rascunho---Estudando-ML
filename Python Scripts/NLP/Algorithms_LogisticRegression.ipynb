### Logistic Regression

Também conhecida como logit regression, esta regressão é conhecida para ser útil em problemas principalmente de calssificação.

Isso funciona muito bem, pois nesses problemas ao invés de tentarmos descobrir um valor arbitrario, queremos descobrir a probabilidade de X ser de uma determinada classe.

O termo logístico, refere-se a probabilidade do logarítimo de ímpares.
Onde, esses ímpares são definidos como:

$$
odds = \frac{P(event)}{1 - P(event)}
$$

A ideia então parte do seguinte. Seja:

$$
Pr(y = 1 | X) = p(X)
$$

Onde o preditor $ X \in Reais$ e $y \in [0,1]$. A regressão logística usa a função logística para cumprir isso (uma das muitas possíveis funções):

$$
\log{\frac{P(event)}{1 - P(event)}}= \beta X
$$

Essa função também é representada como:

$$
p(X) = \frac{1}{1-e^{\beta X}}
$$

Ora, quanto maior o expoente, menor será o valor da potência, e maior será o valor da função.
Da mesma forma, quando menor o expoente, maior será o valor da potência, e menor o valor da função.



### Estimação de parâmetros

Como se sabe no ML, o objetivo é estimar parâmetros $\beta$ reduzindo erros de um modelo para fazer predições.
A regressão logística usa a função <b>the maximum likehood</b> para estimar seus parâmetros. A ideia básica é maximizar os acertos.

Ela funciona da seguinte forma:

- Considere as amostras divididas e rotuladas em 0 e 1.
- Para as amostras rotuladas como 1, é estimado um $\hat \beta$ de forma que $p(X)$ é o mais próximo possível de 1.
- Para as amostras rotuladas como 0, é estimado um $\hat \beta$ de forma que $p(X)$ seja o mais próximo possível de 0. Ou seja, $1 - p(X)$ deve ser o mais próximo possível de 1.

$$
L(\hat \beta) = \prod_{sample \in y_{i}=1}{p(x_{i})} \prod_{sample \in y_{i}=0}{1 - p(x_{i})}
$$

Aqui temos a chamada maximum likehood function:
$$
L(\hat \beta) = \prod_{sample}{ (p(x_{i}))^{y_{i}} (1 - p(x_{i}))^{1 - y_{i}}}
$$

E em seguida, temos o logaritmo da maximum likehood function, forma mais usada para estimação da logit.

$$
L(\hat \beta) = \sum_{sample}{y_{i}log(p(x_{i})) + (1 - y_{i})log(1 - p(x_{i}))}
$$

A forma simplificada (Da pra simplificar no papel) é essa:

$$
L(\hat \beta) = \sum_{sample}{y_{i}\beta x_{i} - log(1 + e^{\beta x_{i}})}
$$

Esse tipo de equação de logaritmos e exponenciais são ditas <b>equações trancedentais</b> e não podem ser computadas exatamente. Para isso usamos métodos numéricos como "Newton Raphson" (Seccant, Moellers, são exemplos de outros métodos).

$$
\hat \beta_{t+1} = \hat \beta_{t} - \frac{\nabla_{\beta} l(\beta_{t})}{\nabla_{\beta \beta} l(\beta_{t})}
$$

Basicamente em cima temos o gradiente da maximum likehood, e em baixo temos a derivada segunda da mesma função. O termo referente ao denominador é a matrix hessiana. A forma final matricial dessa equação é dada por:

$$
\hat \beta_{t+1} = \hat \beta_{t} + (X^{t}W^{t}X)^{-1}X(Y - P(X)^{t})
$$

$\beta_{t}$ convergirá para $\beta$.
