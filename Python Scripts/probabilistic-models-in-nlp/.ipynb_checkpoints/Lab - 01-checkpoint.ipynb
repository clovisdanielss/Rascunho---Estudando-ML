### Distância Mínima de Edição

O problema de adição de distância é um problema que pode ser facilmente resolvido com programação dinâmica através da abordagem do problema do empacotamento. 

A idea do problema consiste na seguinte formulação "Dadas operações, inserir, trocar e deletar, qual o mínimo de operações necessárias para transformar uma palavra em outra ?"

Considerando uma matriz m+1,n+1 onde a primeira linha e a primeira coluna equivale a uma string em branco, temos:


Onde matriz[i,j] é o custo de transformar a substring :i da primeira palavra na substring :j da segunda palavra. Ora, para transformar "p" em "", temos matriz[1,0] que tem custo 1, pois uma operação de deleção resolve. Para transformar "pla" em "", temos matriz[3,0] que tem custo 3, pois são três operação de deleção.

Para as demais casas seguir o seguinte raciocinio:
O custo de transformar "p" em "s" teremos a seguintes opções
"p" -> "s" de custo 2
"p" -> "" -> "s" de custo 2
"p" -> "ps" -> "s" de custo 2

A ideia, é selecionar o menor custo nesse caso temos um empate. Mas generalizando temos o seguite:

$d[i,j] = min(d[i-1,j] +1, d[i,j-1] +1, d[i-1,j-1] + 2)$

Onde 1 é o valor da remoção ou inserção e 2 da troca e a célula d[i-1,j],d[i,j-1],d[i-1,j-1] representa o cumulativo do resto do caminho escolhido até então.

```python
import numpy as np

def distanciaEdicao(palavraA, palavraB):
    matriz  = np.zeros((len(palavraA) + 1, len(palavraB) + 1))
    matriz[0] = np.array([ele for ele in range(matriz.shape[1])])
    matriz[:,0] = np.array([ele for ele in range(matriz.shape[0])])
    for i in range(1,matriz.shape[0]):
        for j in range(1,matriz.shape[1]):
            top = matriz[i-1,j]       +  (1 if palavraA[i-2] != palavraB[j-1] else 0) #valor do salto de linha (deletar | inserir)
            left = matriz[i,j-1]      +  (1 if palavraA[i-1] != palavraB[j-2] else 0) #valor do salto de linha (deletar | inserir)
            topLeft = matriz[i-1,j-1] +  (2 if palavraA[i-2] != palavraB[j-2] else 0) #valor do salto de 2 linhas (trocar)
            if top <= topLeft and top <= left:
                matriz[i,j] = top
            elif left <= top and left <= topLeft:
                matriz[i,j] = left
            elif topLeft <= top and topLeft <= left:
                matriz[i,j] = topLeft
    return matriz

distanciaEdicao("teste","testar")
```
